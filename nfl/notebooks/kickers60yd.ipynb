```python
years = range(1999, 2023)
```

```python
columns = ['kicker_player_name', 'kick_distance', 'play_type', 'field_goal_result']
```

```python
# Group by kicker and count number of 60+ yard field goals
df = df.groupby('kicker_player_name').count().reset_index()
```
