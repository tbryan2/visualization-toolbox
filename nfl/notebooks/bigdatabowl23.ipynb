```python
import imageio
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.animation as animation
from IPython.display import display, HTML
```

```python
def plot_play(play_data):
    fig, ax = plt.subplots(figsize=(12, 5.33))
    plt.xlim(0, 120)
    plt.ylim(0, 53.3)
    plt.title('Play Visualization')
    plt.xlabel('X')
    plt.ylabel('Y')

    # Function to update each frame
    def update(frame):
        ax.clear()
        plt.xlim(0, 120)
        plt.ylim(0, 53.3)
        plt.title('Play Visualization')
        plt.xlabel('X')
        plt.ylabel('Y')
        current_frame = play_data[play_data['frameId'] == frame]
        for idx, row in current_frame.iterrows():
            if row['nflId'] == 'NA':
                ax.scatter(row['x'], row['y'], c='brown', s=100) # Ball
            else:
                ax.scatter(row['x'], row['y'], c='blue' if row['team'] == 'HOU' else 'red')

    ani = animation.FuncAnimation(fig, update, frames=play_data['frameId'].nunique(), interval=100, repeat=False)
    display(HTML(ani.to_jshtml()))
```

```python
df = pd.read_csv('../data/nfl-big-data-bowl-2023/week3.csv')
```

```python
play_data = df[df['playId'] == 54]
plot_play(play_data)
```
